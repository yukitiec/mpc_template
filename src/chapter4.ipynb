# 定数

```python
# ロトカ・ヴォルテラモデルのパラメータ
a = 0.06
b = 0.06
c = 0.024
d = 0.06
e = 0.06
f = 0.012

# 予測ホライズン
K = 30

nu = 1  # 制御変数の次元
nx = 2  # 状態変数の次元
total = nx * (K + 1) + nu * K  # 最適化変数の次元

# コスト関数の重み
Q = casadi.diag([1, 1])
Q_f = casadi.diag([1, 1])
R = casadi.diag([0.05])

# 制約条件
x_lb = [0, 0]
x_ub = [np.inf, np.inf]
u_lb = [0]
u_ub = [1]

# 目標値
x_ref = casadi.DM([1, 1])
u_ref = casadi.DM([0])
```

# 状態方程式

```python
def make_F():  # update function.
    states = casadi.SX.sym("states", nx)
    ctrls = casadi.SX.sym("ctrls", nu)

    x = states[0]
    y = states[1]
    u = ctrls[0]

    x_next = (1 + a) * x - b * x * y - c * x * u
    y_next = (1 - d) * y + e * x * y - f * y * u

    states_next = casadi.vertcat(x_next, y_next)

    F = casadi.Function("F", [states, ctrls], [states_next], ["x", "u"], ["x_next"])
    return F
```

# サイクル現象の確認

```python
t_span = [0, 200]
t_eval = np.arange(*t_span)

x_init = casadi.DM([0.5, 0.7])  # 初期値

F = make_F()

X = [x_init]
x_current = x_init
for t in t_eval:
    x_current = F(x=x_current, u=0)["x_next"]
    X.append(x_current)

X.pop()
X = np.array(X).reshape(t_eval.size, nx)

plt.figure(figsize=(12, 4))

plt.subplot(1, 1, 1)
for k in range(nx):
    plt.plot(t_eval, X[:, k], label=f"x_{k}")
plt.legend()
plt.xlabel("Time")
plt.ylabel("State")

plt.savefig("images/chap4_mpc_no_control.pdf")
plt.show()
```

# コスト関数

```python
def compute_stage_cost(x, u):
    x_diff = x - x_ref
    u_diff = u - u_ref
    cost = (casadi.dot(Q @ x_diff, x_diff) + casadi.dot(R @ u_diff, u_diff)) / 2
    return cost


def compute_terminal_cost(x):
    x_diff = x - x_ref
    cost = casadi.dot(Q_f @ x_diff, x_diff) / 2
    return cost
```

# 最適化問題

```python
def make_nlp():
    # update function.

    F = make_F()


    # parameter setting

    U = [
        casadi.SX.sym(f"u_{k}", nu) for k in range(K)
    ]  # set the variables for the receding horizon.

    X = [
        casadi.SX.sym(f"x_{k}", nx) for k in range(K + 1)
    ]  # include the current state by +1.

    G = []


    J = 0


    # cost calculation

    for k in range(K):

        J += compute_stage_cost(X[k], U[k])

        eq = X[k + 1] - F(x=X[k], u=U[k])["x_next"]

        G.append(eq)

    J += compute_terminal_cost(X[-1])


    # option setting for optimization

    option = {"print_time": False, "ipopt": {"print_level": 0}}
    # nonlinear optimization setting, state, objective function, constraint.

    nlp = {"x": casadi.vertcat(*X, *U), "f": J, "g": casadi.vertcat(*G)}
    # solve the problem with Ipopt

    S = casadi.nlpsol("S", "ipopt", nlp, option)

    return S
```

# 制御入力の計算

```python
def compute_optimal_control(S, x_init, x0):
    # DX->numpy array->unpack->list

    x_init = x_init.full().ravel().tolist()

    # set the lower and upper bound of the state and control.

    lbx = x_init + x_lb * K + u_lb * K  # list

    ubx = x_init + x_ub * K + u_ub * K

    lbg = [0] * nx * K  # equal constraint.

    ubg = [0] * nx * K


    res = S(lbx=lbx, ubx=ubx, lbg=lbg, ubg=ubg, x0=x0)

    offset = nx * (K + 1)

    x0 = res["x"]

    u_opt = x0[offset : offset + nu]
    return u_opt, x0
```

# MPC

```python
S = make_nlp()
```

```python
t_span = [0, 200]
t_eval = np.arange(*t_span, 1)

x_init = casadi.DM([0.5, 0.7])  # 初期値
x0 = casadi.DM.zeros(total)

F = make_F()

X = [x_init]
U = []
x_current = x_init
for t in t_eval:
    u_opt, x0 = compute_optimal_control(S, x_current, x0)
    x_current = F(x=x_current, u=u_opt)["x_next"]
    X.append(x_current)
    U.append(u_opt)
```

# 結果の可視化

```python
X.pop()
X = np.array(X).reshape(t_eval.size, nx)
U = np.array(U).reshape(t_eval.size, nu)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
for k in range(nx):
    plt.plot(t_eval, X[:, k], label=f"x_{k}")
plt.legend()
plt.xlabel("Time")
plt.ylabel("State")

plt.subplot(1, 2, 2)
for k in range(nu):
    plt.step(t_eval, U[:, k], linestyle="--", label=f"u_{k}")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Control")

plt.savefig("images/chap4_mpc.png")
plt.show()
```
